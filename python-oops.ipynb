# BankAccount

```python
class BankAcount:
    
    def getAccount(self,aname,acct_no,acct_type,balance):
        self.aname = aname
        self.acct_no = acct_no
        self.acct_type = acct_type
        self.balance = balance
        
    def displayAccount(self):
        print("\n Account Holder Name : ", aname)
        print("\n Account Number : ", acct_no)
        print("\n Account Type : ", acct_type)
        print("\n Account Balance : ", balance)
        
        
    def depositAccount(self,amount):
        self.balance = balance
        if self.balance <0:
            print('\n Insuffient Balance')
        else:
            tot_balance = (self.balance + amount)
            print('\n After Deposit your Total Amount :', tot_balance)
            
    def withdrawlAccount(self,amount):
        self.balance = balance
        if (self.balance-amount) <0:
            print('\n Fund is not enough to your Current balance ')
        else:
            self.balance = (self.balance - amount)
            print('\n After Withdrawl your Total Amount :', self.balance)
            
    
#main function
if __name__ == '__main__':
    
    ch = 0
    while(ch!=5):       #Menu driven program
        
        text = ''' 
                  Welcome To TestCorporate Bank Account System 
            
        1. Provide Account Information :
        2. Deposit Amount:
        3. Withdrawl Amount:
        4. Display Account Information:
        5  Exit from Service:
        '''
        
        print(text)
        
        ch = int(input('\n enter you choice'))
        ba = BankAcount() # object of bank account class 
        
        
        if ch==1:
            aname = input('\n account holder name :')
            acct_no = input('\n Enter account number :')
            acct_type = input('\n Enter account type :')
            balance = int(input('\n Enter balance :'))
            ba.getAccount(aname,acct_no,acct_type,balance)
              
        if ch==2:
            amt = int(input(f'\n Enter amount that you want to deposite from your bank account {acct_no}:'))
            ba.depositAccount(amt)
                
        if ch==3:
            amt = int(input(f'\n Enter amount that you want to withdraw from your bank account {acct_no} :'))
            ba.withdrawlAccount(amt)

        if ch==4:
            ba.displayAccount()
            
        if ch==5:
            print("\n oops exit from program ...")
            exit(1)
```

```python
class Staff:
    
    def getStaff(self,name,code):
        self.name = name
        self.code =code
        
#     def displayStaff(self):
#         print(f'\n name : {self.name} and code : {self.code}')
        
class Teacher(Staff):
    
    def getTeacher(self,sub,pub):
        self.sub = sub
        self.pub = pub
        
    def displayTeacher(self):
        print(f'\n Name : {self.name} and Code : {self.code}')
        print(f'\n Subject : {self.sub} and Publication : {self.pub}')
        
class Typist(Staff):
    
    def getTypist(self,speed):
        self.speed = speed
        
    def displayTypist(self):
        print(f'\n Name : {self.name} and Code : {self.code}')
        print(f'\n Typist Speed : {self.speed}')
        
class Officer(Staff):
    
    def getOfficer(self,grade):
        self.grade = grade
        
    def displayOfficer(self):
        print(f'\n Name : {self.name} and Code : {self.code}')
        print(f'\n Officer Grade: {self.grade}')
        
class Regular(Typist):
    
    def getRegular(self,salary):
        self.salary = salary
        
    def displayRegular(self):
        print(f'\n Regular Employee Salary: {self.salary}')
    
    
class Casual(Typist):
    
    def getCasual(self,dwages):
        self.dwages = dwages
        
    def displayCasual(self):
        print(f'\n Casual Employee Daily Wages: {self.dwages}')
        
        
if __name__ == '__main__':
    
    ch=0
    while(ch!=5):
        
        text = '''
        1. Get Teacher Info:
        2. Get Officer Info:
        3. Get Regular Employee Info:
        4. Get Casual Employee Info:
        5. Exit from program
        '''
        
        print(text)
        ch = int(input('\n Enter your choice'))
        ta = Teacher()
        if(ch==1):
            name = input('\n enter teacher name :')
            code = input('\n enter teacher code :')
            subject = input('\n type subject :')
            publication = input('\n type publication name :')
            ta.getTeacher(subject,publication)
            ta.getStaff(name,code)
            ta.displayTeacher()
            
        of = Officer()
        if(ch==2):
            name = input('\n enter officer name :')
            code = input('\n enter officer code :')
            grade = input('\n type officer grade :')
            of.getOfficer(grade)
            of.getStaff(name,code)
            of.displayOfficer()
            
        r = Regular()
        if(ch==3):
            name = input('\n enter Regular employee name :')
            code = input('\n enter Regular employee code :')
            salary = input('\n salery  :')
            speed = int(input('\n enter typing speed'))
            r.getStaff(name,code)
            r.getRegular(salary)
            r.getTypist(speed)
            r.displayTypist()
            r.displayRegular()
            
        c = Casual()
        if(ch==4):
            name = input('\n enter Casual employee name :')
            code = input('\n enter Casual employee code :')
            dwages = int(input('\n Daily wages :'))
            speed = int(input('\n enter typing speed'))
            c.getStaff(name,code)
            c.getCasual(dwages)
            c.getTypist(speed)
            c.displayTypist()
            c.displayCasual()
            
        if(ch==5):
            print('\n Exit from program')
            exit(1)
        
```

# Define Constructor and Destructor in python

```python
class getSum:
    def __init__(self,a,b):  # constuctor
        self.a =a
        self.b =b
        
    def showMe(self):
        print(f'\n add : {self.a + self.b}')
        
    def __del__(self):       # destuctor
        class_name = self.__class__.__name__
        print("\n Object Deleted for :", class_name)
        
if __name__ == '__main__':
    
    try:
        a1 = getSum(5,6)
        print(type(a1))
        a1.showMe()
        
        del a1  #delete object
        
        if (a1.showMe()):  #try to show object again and get error, so applied try-except block here
            pass
    
    except Exception:
        print('\n Once Deleted can not create object again, until remove destructor')
```

# Method Overriding ---=) Polymorphism

```python
class Kali:
    
    def advantage(self):
        print('\n Extended support')
        print('\n Widely used')
    
    def disadvantage(self):
        print('\n slow performance')
        print('\n Max memory consumed')
        
class Parrot:
    
    def advantage(self):
        print('\n Faster Perfomance')
        print('\n Min memory occupied')
        
    def disadvantage(self):
        print('\n Less support')
        print('\n Frequently used')
        
k1 = Kali()
p1 = Parrot()
print('\n======Kali===========')
k1.advantage()
k1.disadvantage()
print('\n======Parrot===========')
p1.advantage()
p1.disadvantage()
```

# Operator Overloading using Dunder / Double Underscore                                      / Magic method of python

```python
class A:
    
    def __init__(self,a,b):
        self.a = a
        self.b = b
        
    def __add__(self,other):
        return(other.a+self.a, other.b+self.b)
    
    def __sub__(self,other):
        return(other.a-self.a, other.b-self.b)
    
    def __mul__(self,other):
        return(other.a*self.a, other.b*self.b)
    
    def __truediv__(self,other):
        return(other.a/self.a, other.b/self.b)
        
    def __str__(self):
        A("%d %d",self.a,self.b)
        
        
a1=A(5,6)
a2=A(7,8)
print(a1+a2)
print(a2-a1)
print(a2*a1)
print(a2/a1)
```

# Data Encapsuation

```python
'''
    Data Encapsulation ==> Data Hiding ==> 
    
    __(private mode), _(protected mode), no underscore (default public mode)
'''

class dataHide:
    
    def getData(self,_x,_y):
        self._x=_x 
        self._y=_y
        
        if(self._x > self._y):
            self._x += 1
        else:
            self._y += 100
        
    def __showData(self):
        print(self._x)
        print(self._y)
        
d1 = dataHide()
d1.getData(11,12)
d1.__showData()

'''
Note : Though I've declared x and y variables in protected mode, showData()defines it as a protected mode function. 
Therefore It shows below Attribute Error which indicates showData() function is not visible outside of class dataHide.

'''
```
